# Matplotlib and Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
new_df = pd.read_csv('rainfall in india 1901-2015.csv')
```

```python
new_df.head()
```

```python
new_df1 = new_df.groupby('SUBDIVISION').mean()
```

```python
new_df1.head()
```

```python
new_df1.mean()
```

# plt.plot using matplotlib

```python
plt.figure(figsize=(30,10))
plt.xticks(rotation = 90)
plt.plot(new_df1['Jan-Feb'],label = 'January - February', color = 'red', linestyle = 'solid', linewidth =4)
plt.title('Rainfall over years')
plt.xlabel('Regions')
plt.ylabel('Rainfall in cm')
plt.legend(loc='right')
plt.show()
```

```python
plt.figure(figsize=(40,10))
sns.set_theme(style = 'darkgrid', font_scale = 0.8)
plt.xticks(rotation = 90)
plt.plot(new_df1['Jan-Feb'],label='January-February',linestyle = 'solid',color = 'red',linewidth=4)
plt.title('Rainfall over regions of India')
plt.xlabel('Regions')
plt.ylabel('Rainfall in cm')
plt.legend(loc = 'upper center')
plt.plot()
```

# Subplots

```python
plt.figure(figsize=(40,50))
sns.set_theme(style='darkgrid',font_scale = 1.5)
plt.subplots_adjust(wspace = 0.6,hspace = 0.8)
plt.subplot(2,1,1)
plt.xticks(rotation = 90)
plt.plot(new_df1['Jan-Feb'],label = 'January-February', linestyle = 'solid',color = 'green', linewidth = 4)
plt.subplot(2,1,2)
plt.xticks(rotation = 90)
plt.plot(new_df1['Mar-May'],label = 'March-May', linestyle = 'dashed', color = 'black', linewidth = 8)
plt.show()
```

# Pie chart

```python
df_pie = pd.read_csv('district wise rainfall normal.csv')
```

```python
df_pie.head()
```

```python
df_pie[['DISTRICT','ANNUAL']]
```

```python
plt.figure(figsize=(40,50))
palette_color_1 = sns.color_palette('pastel')
palette_color_2 = sns.color_palette('dark')
palette_color_3 = sns.color_palette('tab10')
explode_list=[0.2,0,0,0,0,0,0,0,0,0]
plt.pie(df_pie.iloc[:10,14],labels=df_pie.iloc[:10,1],shadow=False,autopct='%.2f%%',colors=palette_color_1,explode=explode_list)
plt.title('Pie Chart for Rainfall')
plt.show()
```

# Violin Plot

```python
plt.xticks(rotation=90)
sns.violinplot(x=new_df1['YEAR'],y = new_df1['ANNUAL'],style ="event") #some other styles include whitegrid etc.
plt.show()
```

# Bar plot

```python
plt.xticks(rotation=90)
plt.title('Bar Plot for Districts')
sns.barplot(x='DISTRICT', y='Jan-Feb', data=df_pie.iloc[:15,:], palette='magma') #palette: magma, muted, Set2, Set3
plt.show()
```
